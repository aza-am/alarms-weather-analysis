```python
import pandas as pd
```

<h2>Cheking data</h2>

```python
df_merged = pd.read_csv("merged_weather_alarms.csv")

df_isw = pd.read_csv("isw_tfidf.csv")

df_telegram = pd.read_csv("telegram_tfidf.csv")
```

```python
print("df_merged --> ", df_merged.shape)
print("df_isw --> ", df_isw.shape)
print("df_telegram --> ", df_telegram.shape)


print("Duplicates in df_merged datetime --> ", df_merged['datetime'].duplicated().sum())
print("Duplicates in df_isw datetime --> ", df_isw['datetime'].duplicated().sum())
print("Duplicates in df_telegram datetime --> ", df_telegram['datetime'].duplicated().sum())
```

```python
df_merged.head(2)
```

```python
df_isw.head(2)
```

```python
df_telegram.head(2)
```

```python
print(df_merged["datetime"].dtype)

print(df_isw["datetime"].dtype)

print(df_telegram["datetime"].dtype)
```

```python
df_merged['datetime'] = pd.to_datetime(df_merged['datetime'])
df_isw['datetime'] = pd.to_datetime(df_isw['datetime'])
df_telegram['datetime'] = pd.to_datetime(df_telegram['datetime'])

print(df_merged["datetime"].dtype)
print(df_isw["datetime"].dtype)
print(df_telegram["datetime"].dtype)
```

```python
isw_colums = [col for col in df_isw.columns if col != 'datetime']
df_isw[isw_colums] = df_isw[isw_colums].astype('float32')

telegram_colums = [col for col in df_telegram.columns if col != 'datetime']
df_telegram[telegram_colums] = df_telegram[telegram_colums].astype('float32')
```

```python
print(df_isw.dtypes)
print(df_telegram.dtypes)
```

<h3>Merging isw data with weather and events</h3>

```python
df = df_merged.merge(df_isw, on='datetime', how='left')
```

<h4>Cheking again</h4>

```python
df.head(2)
```

```python
df.shape
```

```python
df.isna().sum().sum()
```

```python
isw_colums = [c for c in df.columns if c.startswith('isw_tfidf_')]
df[isw_colums] = df[isw_colums].fillna(0)

df.isna().sum().sum()
```

```python
df.to_csv('100-isw-weather-events.csv', index=False)
```

<h2>Merging telegram</h2>

<h3>Shifting hour</h3>

```python
df_telegram['datetime'] = pd.to_datetime(df_telegram['datetime']) + pd.Timedelta(hours=1)
```

```python
df_telegram.head(2)
```

```python
completed_df = df.merge(df_telegram, on='datetime', how='left')
```
