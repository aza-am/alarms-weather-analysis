```python
tfidf_cols = [str(i) for i in range(100)]
df_isw_tfidf = df_isw_tfidf.rename(
    columns={col: f'isw_tfidf_{col}' for col in tfidf_cols}
)
df_isw_tfidf.columns.tolist()[-5:]
```

```python
# For expanding ISW article to 24 hours
def expand_isw_hours(row):
    hours = pd.date_range(
        start=row['date'],
        periods=24,
        freq='h'
    )
    df_temp = pd.DataFrame({'datetime': hours})
    for col in [c for c in df_isw_tfidf.columns if c.startswith('isw_tfidf_')]:
        df_temp[col] = row[col]
    return df_temp

df_isw_hourly = pd.concat(
    df_isw_tfidf.apply(expand_isw_hours, axis=1).tolist(),
    ignore_index=True
)

df_isw_hourly.shape
```

**MERGE WEATHER AND EVENTS**

```python
df_weather = pd.read_csv(path + 'weather_clean.csv', parse_dates=['datetime'])
df_alarms_hourly = pd.read_csv(path + 'alarms_hourly.csv', parse_dates=['datetime'])
```

```python
df_weather.shape
```

```python
df_alarms_hourly.shape
```

```python
df_weather[['datetime', 'region_id']].dtypes
```

```python
df_alarms_hourly[['datetime', 'region_id']].dtypes
```

```python
# LEFT JOIN(df_weather is our main matrix, df_alarms_hourly contains only the hours when an alarm was)
df_merged = df_weather.merge(
    df_alarms_hourly[['datetime', 'region_id', 'alarm']],
    on=['datetime', 'region_id'],
    how='left'
)

# Where there was no alarm, NaN appears, so we replace with 0
df_merged['alarm'] = df_merged['alarm'].fillna(0).astype(int)
```

```python
df_merged.shape
```

```python
(df_merged['alarm']==0).sum()
```

```python
(df_merged['alarm']==1).sum()
```

```python
df_merged.isnull().sum().sum()
```

```python
# For calculating the number of alarms in all regions for the current hour
df_merged['regions_in_alarm_now'] = df_merged.groupby('datetime')['alarm'].transform('sum')
```

```python
df_merged.head()
```

```python
# For sorting the dataframe by region and time to prevent mixing data from different regions in the rolling window
df_merged = df_merged.sort_values(by=['region_id', 'datetime']).reset_index(drop=True)

# For calculating the rolling sum of alarms over the last 24 hours (shifting by 1 hour to exclude the current one)
df_merged['alarms_last_24h'] = df_merged.groupby('region_id')['alarm'].transform(
    lambda x: x.rolling(window=24, min_periods=1).sum().shift(1)
).fillna(0)
```

```python
df_merged.head()
```

```python
# For extracting a time-based feature(1 if it's night from 22:00 - 06:00), 0 otherwise)
df_merged['is_night'] = df_merged['datetime'].dt.hour.apply(lambda x: 1 if (x >= 23 or x <= 6) else 0)
```

```python
df_merged.head()
```

```python
# For creating a lag feature(whether there was an alarm in this region exactly 24 hours ago)
df_merged['alarm_24h_ago'] = df_merged.groupby('region_id')['alarm'].shift(24).fillna(0)
```

```python
df_merged['alarm_24h_ago'].value_counts()
```
